# **1.9. Naive Bayes**

```python
import seaborn as sb

d = sb.load_dataset('iris')
```

```python
d.head()
```

Split Depenedent(y) and Independen(x)

```python
x = d[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']]
y = d['species']
```

*train test split (from sklearn.model_selection) *

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state=10)
```

# **Bernoulli Naive Bayes**

```python
from sklearn.naive_bayes import BernoulliNB
from sklearn.metrics import accuracy_score
model = BernoulliNB()
model.fit(X_train, y_train)
```

```python
bPread = model.predict(X_test)
print(accuracy_score(y_test, bPread))
```

```python
from sklearn.metrics import confusion_matrix
print(confusion_matrix(y_test, bPread))
print(7/30)
```

# **Gaussian Naive Bayes**

```python
from sklearn.naive_bayes import GaussianNB
model2 = GaussianNB()
model2.fit(X_train, y_train)
```

```python
gPred = model2.predict(X_test)
print(accuracy_score(y_test, gPred))
```

```python
print(confusion_matrix(y_test, gPred))
print(30/30)
```

# **Multinomial Naive Bayes**

```python
from sklearn.naive_bayes import MultinomialNB
model3 = MultinomialNB()
model3.fit(X_train, y_train)
```

```python
mPred = model3.predict(X_test)
print(accuracy_score(y_test, mPred))
```

```python
print(confusion_matrix(y_test, mPred))
print(18/30)
```
